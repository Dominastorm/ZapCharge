```python
import pandas as pd
import folium
import random
```

```python
json_file_path = "../../data/charger_map_data.json"
df = pd.read_json(json_file_path)
```

```python
# type_2 = AC
# type_3 = ONLY DC FAST CHARGERS
```

```python
df
```

```python
num_rows = 2259

df_1 = pd.DataFrame(columns = ['idx', 'charger_type', 'daily_usage_2', 'peak_usage_2', 'daily_usage_3', 'peak_usage_3', 'type_2_rating', 'type_3_rating', 'type_2_nos', 'type_3_nos', 'energy_usage', 'uptime'])
daily_usage_range = [] # hours
for i in range(1, 24):
    for j in range(0, 10):
        daily_usage_range.append(i + 0.1 * j)

type_2_rating_range = [4, 8, 10, 15, 20, 22] # kW
type_3_rating_range = [200, 250, 300,350, 400, 450] # kW
uptime_range = [x for x in range(0, 200, 5)] # hours


for i in range(num_rows):
    idx = i
    charger_type = df['charger_type'][i]

    type_2_rating = random.choice(type_2_rating_range)
    type_3_rating = random.choice(type_3_rating_range)
    type_2_nos = random.randint(1,6)
    type_3_nos = random.randint(1,3)

    daily_usage_2 = round(random.choice(daily_usage_range) * type_2_nos, 2)
    peak_usage_2 = round( min(24, daily_usage_2 * (1 + 0.1 * random.randint(0, 15) )), 2) * type_2_nos

    daily_usage_3 = round(random.choice(daily_usage_range) * type_3_nos, 2)
    peak_usage_3 = round( min(24, daily_usage_3 * (1 + 0.1 * random.randint(0, 15) )), 2) * type_3_nos

    energy_usage = round(type_2_nos * daily_usage_2 + type_3_nos * daily_usage_3, 2)
    uptime = random.choice(uptime_range)
    df_1.loc[i] = [idx, charger_type, daily_usage_2, peak_usage_2, daily_usage_3, peak_usage_3, type_2_rating, type_3_rating, type_2_nos, type_3_nos, energy_usage, uptime]

df_1
```

```python
df_1.to_csv('../../data/charger_consumption_data.csv')
```

```python
# 1. Plot stations whose working exceeds the specified threshold
```

```python
daily_usage_2_limit , peak_usage_2_limit = 20, 22
daily_usage_3_limit, peak_usage_3_limit =  15, 18
```

```python
coordinates_1 = []
for i in range(len(df_1)):
    if (df_1['daily_usage_2'][i] / df_1['type_2_nos'][i]) >= daily_usage_2_limit or (df_1['peak_usage_2'][i] / df_1['type_2_nos'][i]) >= peak_usage_2_limit:
        coordinates_1.append([df['latitude'][i], df['longitude'][i]])
```

```python
coordinates_2 = []
for i in range(len(df_1)):
    if (df_1['daily_usage_3'][i] / df_1['type_3_nos'][i]) >= daily_usage_3_limit or (df_1['peak_usage_3'][i] / df_1['type_3_nos'][i]) >= peak_usage_3_limit:
        coordinates_2.append([df['latitude'][i], df['longitude'][i]])
```

```python
map = folium.Map(location=city_coords, zoom_start=10, control_scale = True)
for latitude, longitude in coordinates_1:
    if latitude and longitude:
        folium.Marker(location=(latitude, longitude), icon=folium.Icon()).add_to(map)
map
```
